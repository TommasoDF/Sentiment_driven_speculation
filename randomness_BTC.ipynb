```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.stats import levy_stable
```

```python
# Getting btc data
start = "2018-1-1"
end = "2023-4-30" #df.index[-1]
data = yf.download('AAPL', start=start, end=end)
```

```python
#Computing log returns
data['log_ret'] = np.log(data.Close/data.Open)
```

```python
#Fit an alpha stable distribution
alpha, beta, loc, scale = levy_stable._fitstart(data.log_ret)
params = levy_stable.fit(data.log_ret, floc = 0, fscale = scale)
```
